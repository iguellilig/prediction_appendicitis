```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import joblib

# random seed
seed = 42
```

```python
# Read original dataset
data_df = pd.read_csv("data/data.csv")
data_df.sample(frac=1, random_state=seed)
#data_df = data_df[np.isfinite(X).all(1)]
#data_df.fillna(method = 'ffill', inplace = True)
#data_df.fillna(method = 'bfill', inplace = True)
data_df = data_df.dropna(subset=["Diagnosis"])
data_df.fillna(-1, inplace = True)
#data_df = data_df.dropna()
# selecting features and target data
X = data_df[['Age', 
             'BMI', 
             'Sex', 
             'Height',
             'Weight',
             'Appendix_Diameter',
             'Migratory_Pain', 
             'Lower_Right_Abd_Pain',
             'Contralateral_Rebound_Tenderness',
             'Coughing_Pain',
             'Loss_of_Appetite',
             'Body_Temperature',
             'WBC_Count',
             'Neutrophil_Percentage',
             'Segmented_Neutrophils',
             'Neutrophilia',
             'RBC_Count',
             'Hemoglobin',
             'RDW',
             'Thrombocyte_Count',
             'Ketones_in_Urine',
             'RBC_in_Urine',
             'WBC_in_Urine',
             'CRP',
             'Dysuria',
             'Stool',
             'Peritonitis',
             'Psoas_Sign'
            ]]
             
y = data_df[['Diagnosis']]

X_val_map = [('Sex', {'male':0, 'female':1}), 
                    ('Migratory_Pain', {'yes': 2, 'no':3}), 
                    ('Lower_Right_Abd_Pain', {'yes': 2, 'no':3}),
                    ('Contralateral_Rebound_Tenderness', {'yes': 2, 'no':3}),
                    ('Coughing_Pain', {'yes': 2, 'no':3}),
                    ('Loss_of_Appetite', {'yes': 2, 'no':3}),
                    ('Neutrophilia', {'yes': 2, 'no':3}),
                    ('Ketones_in_Urine', {'no': 3, '+':4, '++':5, '+++':6}),
                    ('RBC_in_Urine', {'no': 3, '+':4, '++':5, '+++':6}),
                    ('WBC_in_Urine', {'no': 3, '+':4, '++':5, '+++':6}),
                    ('Dysuria', {'yes': 2, 'no':3}),
                    ('Stool', {'normal': 7, 'constipation':8, 'diarrhea':9, 'constipation, diarrhea':10}),
                    ('Peritonitis', {'no': 3, 'local':11, 'generalized':12}),
                    ('Psoas_Sign', {'yes': 2, 'no':3})]
for column, themap in X_val_map:
    X[column] = X[column].replace(themap)
    #X[column].replace(-1, np.nan,inplace=True)

Y_val_map = [('Diagnosis', {'appendicitis':1, 'no appendicitis': 0})]
for column, themap in Y_val_map:
    y[column] = y[column].replace(themap)

X = X.astype(float)
X = X.round(3)
'''X = pd.get_dummies(X, columns=['Sex',
                               'Migratory_Pain',
                               'Lower_Right_Abd_Pain',
                               'Contralateral_Rebound_Tenderness',
                               'Coughing_Pain',
                               'Loss_of_Appetite',
                               'Ketones_in_Urine',
                               'RBC_in_Urine',
                               'WBC_in_Urine',
                               'Dysuria',
                               'Stool',
                               'Peritonitis',
                               'Psoas_Sign',
                               'Segmented_Neutrophils',
                               'Neutrophilia'], drop_first=True)'''
'''print (X)

y = pd.get_dummies(y, columns=['Diagnosis'], drop_first=True)

print(y)'''
#X.fillna(method = 'ffill', inplace = True)
#y.fillna(method = 'ffill', inplace = True)

#X = X[np.isfinite(X).all(1)]
#y = y[np.isfinite(y).all(1)]


#print(X)
#print(y)

# split data into train and test sets
# 70% training and 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=seed, stratify=y)

print(y_test)
```

```python
# create an instance of the random forest classifier
clf = RandomForestClassifier(n_estimators=100)

# train the classifier on the training data
clf.fit(X_train, y_train)

# predict on the test set
y_pred = clf.predict(X_test)
print(y)
print(y_pred)

# calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")  # Accuracy: 0.91

# save the model to disk
joblib.dump(clf, "rf_model.sav")
```
